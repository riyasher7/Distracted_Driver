```r
library(ggplot2)
library(dplyr)
```

```r
data <- read.csv("goldm1.csv")
data$Date <- as.Date(data$Date, format="%Y-%m-%d")
```

```r
plot1 <- ggplot(data, aes(x = Date, y = Close)) +
  geom_line(color = "blue") +
  labs(title = "Closing Prices Over Time", x = "Date", y = "Closing Price") +
  theme_minimal()
```

```r
plot2 <- ggplot(data, aes(x = Date, y = Volume)) +
  geom_line(color = "darkgreen") +
  labs(title = "Volume Traded Over Time", x = "Date", y = "Volume") +
  theme_minimal()
```

```r
plot3 <- ggplot(data, aes(x = High, y = Low)) +
  geom_point(alpha = 0.5, color = "purple") +
  labs(title = "High vs Low Prices", x = "High Price", y = "Low Price") +
  theme_minimal()
```

```r
# Assuming 'data' is your dataframe and 'Close' is the column you want to plot
plot4 <- ggplot(data, aes(x = as.numeric(Close))) +  # Convert 'Close' to numeric
  geom_histogram(binwidth = 5, fill = "orange", color = "black", alpha = 0.7) +
  labs(title = "Distribution of Closing Prices", x = "Closing Price", y = "Frequency") +
  theme_minimal()
```

```r
data <- data %>% mutate(YearMonth = format(Date, "%Y-%m"))
plot5 <- ggplot(data, aes(x = YearMonth, y = Close)) +
  geom_boxplot(fill = "skyblue", color = "darkblue") +
  labs(title = "Monthly Trends in Closing Prices", x = "Year-Month", y = "Closing Price") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  theme_minimal()
```

```r
# Display plots in Colab
print(plot1)  # Line Plot: Closing Prices Over Time
```
